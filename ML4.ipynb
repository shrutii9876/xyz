```python
x_values=[]
y_values=[]
```

```python
def gradient_descent(starting_point, learning_rate, num_iterations):

    x=starting_point

    for i in range (num_iterations):
        x_values.append(x)
        y_values.append((x+3)**2)
        gradient = 2 * (x+3)
        x = x - learning_rate * gradient

        print(f"Iteration {i+1} : x={x}, y={(x+3)**2}")

    return x
```

```python
starting_point = 2
learning_rate = 0.1
num_iterations = 20
```

```python
final_x=gradient_descent(starting_point, learning_rate,num_iterations)
print(f"Local minimum found at x: {final_x}, y={(final_x+3)**2}")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_values, y_values, 'ro-')
plt.title('Gradient Descent Visualization for y = (x + 3)^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
